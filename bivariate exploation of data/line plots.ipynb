### in a line plot, only one point is plotted for every unique x-value or bin of x-values (like a histogram).

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fuel_econ=pd.read_csv('fuel-econ.csv')
```

**Example 1**

```python
plt.errorbar(data=fuel_econ,x='displ',y='comb')
plt.xticks(rotation=15)
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

the function above plots all the data points as a line, connecting values from the first row of the df to the last row
